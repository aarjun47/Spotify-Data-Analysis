```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing data
dtracks = pd.read_csv('tracks.csv')
```

```python
#checking for null values
pd.isnull(dtracks).sum()
```

```python
#sorting data in basis of popularity
sorteddf = dtracks.sort_values('popularity',ascending = True).head(10)
```

```python
ppl = dtracks.query('popularity > 90', inplace = False).sort_values('popularity',ascending = False)
ppl[:10]
```

```python
ppl.head(10)
```

```python
# making the release date as index
dtracks.set_index("release_date",inplace = True)
dtracks.index=pd.to_datetime(dtracks.index)
dtracks.head()
```

```python
dtracks.info()
```

```python
dtracks.release_date
```

```python
# converting the duration to seconds
dtracks["duration"]=dtracks["duration_ms"].apply(lambda x: round(x/1000))
dtracks.drop("duration_ms",inplace=True , axis=1)
```

```python
dtracks
```

```python
#conducting correlation
corrr = dtracks.drop(["key","mode","explicit"],axis=1).corr(method = "pearson")
```

```python
corrr
```

```python
# heatmap
plt.figure(figsize =(14,6))

ht = sns.heatmap(corrr,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidth=1,linecolor="Black")
```

```python
sample_data = dtracks.sample(int(0.004*len(dtracks)))
```

```python
sample_data
```

```python
# comparing loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data = sample_data , x='energy' ,y='loudness',color='c').set(title = 'loudness vs energy')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_data , x='acousticness' ,y='popularity',color='Blue').set(title = 'loudness vs energy')
```

```python
dtracks['dates']=dtracks.index.get_level_values('release_date')
dtracks.dates=pd.to_datetime(dtracks.dates)
```

```python
years=dtracks.dates.dt.year
```

```python
# Per year distrubution of songs
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title='no of songs per year')
```

```python
# showing how the duration of songs changes according to the year
totaldura=dtracks.duration
fig_dim = (18,7)
fig,ax=plt.subplots(figsize=fig_dim)
fig=sns.barplot(x=years,y=totaldura,ax = ax,errwidth=False).set(title='Year vs Duration')
plt.xticks(rotation=90)
```

```python
totaldura=dtracks.duration
fig_dim = (10,5)
fig, ax=plt.subplots(figsize=fig_dim)
fig=sns.lineplot(x=years,y=totaldura,ax = ax).set(title='Year vs Duration')
```
